## 7 - Manipule os dados a seguir:

### Criar coluna "Total" somando o valor de todos os ativos por investidor

```python
cadastro_invest['Total'] = cadastro_invest['Ativo01'] + cadastro_invest['Ativo02'] + cadastro_invest['Ativo03'] + cadastro_invest['Ativo04'] + cadastro_invest['Ativo05'] + cadastro_invest['Ativo06'] + cadastro_invest['Ativo07'] 
```

```python
cadastro_invest
```

### Criar coluna "Media" com a média das notas 1 e 2

```python
cadastro_invest['Media'] = (cadastro_invest['Nota1'] + cadastro_invest['Nota2']) /2
```

```python
cadastro_invest
```

### Criar coluna "Dominio" com apenas o domínio do email (dica: separar por @)

```python
cadastro_invest['Dominio'] = cadastro_invest['Email'].str.split('@').str.get(1)
```

```python
cadastro_invest
```

### Mudar o nome da coluna "Salário" para "Renda" e tratar a coluna (sem caracter e tipo numérica)

```python
cadastro_invest.rename(columns={'Salario' : 'Renda'}, inplace=True)
cadastro_invest['Renda'] = cadastro_invest['Renda'].str.split('$').str.get(1)
pd.to_numeric(cadastro_invest['Renda'])
```

```python
cadastro_invest.info()
```

### Mudar os dados da coluna "Sexo" para Masculino e Feminino (tente usar a função map)

```python
cadastro_invest['Sexo'] = cadastro_invest['Sexo'].map({'Fem':'Feminino' , 'Masc':'Masculino'}, na_action=None)
```

```python
cadastro_invest
```

### Transformar os dados da coluna Media em inteiro

```python
cadastro_invest['Media'] = cadastro_invest['Media'].astype(int)
```

```python
cadastro_invest
```

### Criar coluna "NPS" onde o investidor é classificado como promotor se a média das notas é maior ou igual a 80, como neutro se a média estiver entre 50 e 80, e como detrator se for abaixo de 50

```python
cadastro_invest.loc[cadastro_invest['Media'] >= 80, 'NPS'] = 'Promotor'
cadastro_invest.loc[cadastro_invest['Media'] < 80 , 'NPS'] = 'Neutro'
cadastro_invest.loc[cadastro_invest['Media'] <50, 'NPS'] = 'Detrator'
```

```python
cadastro_invest
```
